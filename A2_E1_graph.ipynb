```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,6,50)
```

```python
plt.figure()

plt.plot([0,6.0],[3.0,0], color='blue', alpha = 0.5)
plt.fill_between([0,6.0], [3.0,0], color='blue', alpha=0.25, label='Materia prima M1')

plt.plot([0,4.0],[6.0,0], color='pink', alpha = 0.5)
plt.fill_between([0,4.0], [6.0,0], color='pink', alpha=0.25, label='Materia prima M2')

plt.plot(x,1+x, color='red', alpha = 0.5)
plt.axhline(2)

plt.plot([0,72.0/18.0],[72.0/12.0,0], color='black', alpha = 1)

plt.scatter([3], [1.5], color='red')  # punto Ã³ptimo

plt.xlim(0)
plt.ylim(0)

plt.xlabel('exteriores')
plt.ylabel('interiores')
plt.legend()
```
