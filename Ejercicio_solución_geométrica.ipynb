#### Ejercicio 1

Una compañía fabrica dos tipos de motores pequeños con combustible sólido. Con el motor A gana $ \$  3$ por motor y con el motor B gana $\$ 4$ por motor. Dispone de 80 horas por semana para producirlos y en los procesos se requieren: 4 h para A y solo 2 h para B. Sin embargo, debido a lo peligroso del material usado, son necesarias 5 h de preparación y limpieza para el motopr B y 2 h para A. El tiempo de preparación total disponible es 120 h por semana. Determinar el número de cada clase de motores a producir que dé como resultado la máxima ganancia.   

De manera formal:

#### Variables de decisión:

$x_1 :=$ cantidad de motores A,
$x_2 :=$ cantidad de motores B

#### Función a optimizar:
$$z=3x_1+4x_2$$

Restricciones:

$$4x_1+2x_2\leq 80$$
$$2x_1+5x_2\leq 120$$

Resultante del tiempo producción $P_1$: $2x_1 + 8x_2 \leq 60$

Resultante del tiempo de limpieza y preparación $P_2$: $4x_1 + 4x_2 \leq 60$

No negatividad: $x_1 \geq 0$, $x_2 \geq 0$

```python
import matplotlib.pyplot as plt #Es la librería para graficar
%matplotlib inline
```

```python
 plt.figure()

 # Restricción para P1
 plt.plot([0, 80.0/2.0], [80.0/4.0, 0], color='orange', alpha=0.25)
 plt.fill_between([0, 80.0/2.0], [80.0/4.0, 0], color='orange', alpha=0.25, label='resultante de M1')

 # Restricción para P2
 plt.plot([0, 120.0/5.0], [120.0/2.0, 0], color='blue', alpha=0.125)
 plt.fill_between([0, 120.0/5.0], [120.0/2.0, 0], color='blue', alpha=0.125, label='resultante de M2')

 plt.xlabel('motores A')
 plt.ylabel('motores B')

 plt.scatter([10], [10], color='black')  # punto factible
 plt.plot([4.0*10.0/3+10.0, 0], [0,3.0*10.0/4+10.0], color='black', linestyle='--')

 plt.scatter([20], [10], color='red')  # punto óptimo
 plt.plot([100.0/3,0], [0,25.0], color='red', linestyle='-.')


 plt.xlim(0)
 plt.ylim(0)

 plt.legend()
 plt.show()
```
