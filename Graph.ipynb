```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return np.exp(x)
```

```python
f(1)
```

```python
x = np.linspace(-5,5,100)
```

```python
plt.plot(x,f(x))
```

```python
plt.plot(x,f(x), color="purple")
plt.title("Gráfica de la función exponencial en [-5,5]")
plt.xlabel("x")
plt.ylabel("y")
```
